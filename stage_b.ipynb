## Question 13

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
energy_df = pd.read_csv('energydata_complete.csv')
energy_df = energy_df.drop(['date', 'lights'], axis=1)

# Normalizing the data
scaler = MinMaxScaler()
scaled = scaler.fit_transform(energy_df)
normalised_df = pd.DataFrame(scaler.fit_transform(energy_df), columns=energy_df.columns)
#features_df = normalised_df.drop(columns=['T2', 'T6'])
features_df = normalised_df
target_variable = normalised_df['Appliances']

# Split data into train and test
x_train, x_test, y_train, y_test = train_test_split(features_df, target_variable,
                                    test_size=0.3, random_state=42)
linear_model = LinearRegression()

linear_model.fit(x_train, y_train)

y_predicted = linear_model.predict(x_test)

from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_predicted)
print(f'Mean absolute error is: {round(mae, 5)}')
```

## Question 12

```python
from sklearn.metrics import r2_score

sns.regplot(x='T2', y='T6', data = energy_df)
r2_score = r2_score(energy_df['T2'], energy_df['T6'])
print(f'R2 value is: {round(r2_score, 2)}')
```

## Question 14

```python
import numpy as np
rss = np.sum(np.square(y_test - y_predicted))
print(f'Residual Sum of Squares is: {round(rss, 5)}')
```

## Question 15

```python
from sklearn.metrics import mean_squared_error
rmse = np.sqrt(mean_squared_error(y_test, y_predicted))
print(f'Root Mean Squared Error is: {round(rmse, 5)}')
```
